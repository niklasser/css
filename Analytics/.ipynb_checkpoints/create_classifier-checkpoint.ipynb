# CSS Erstellen des Klassifikators

```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
from influxdb import InfluxDBClient
from influxdb import DataFrameClient
```

```python
from sklearn.preprocessing import label_binarize
```

Create InfluxDB Connection

```python
client = DataFrameClient(host='localhost', port=8086)
```

```python
client.get_list_database()
```

```python
client.switch_database("training")
```

```python
result = client.query("select * FROM contextdata")
```

```python
df = pd.DataFrame(data=result["contextdata"])
```

```python
df.sample(5)
```

```python
print(len(df))
```

```python
df.dropna()
print(len(df))
```

```python
df = df[["label", "x", "y", "z"]]
```

```python
df.head()
```

```python
df.label.unique()
```

```python
df["label"] = label_binarize(df["label"], df.label.unique())
```

# Visulizing the Data

```python
sns.pairplot(df, hue="label")
plt.show()
```

```python
for cat in df[["x", "y", "z"]]:
    sns.distplot(df[df["label"] == 0][cat], kde=True, color="blue", hist=False)
    sns.distplot(df[df["label"] == 1][cat], kde=True, color="orange", hist=False)
    plt.show()
```

# Classifier

```python
from sklearn.tree import DecisionTreeClassifier
from sklearn.ensemble import RandomForestClassifier
from sklearn.naive_bayes import GaussianNB
from sklearn.svm import SVC

from sklearn.model_selection import cross_validate
```

```python
classifiers = [
    DecisionTreeClassifier(),
    GaussianNB(),
    RandomForestClassifier(n_estimators=5),
    SVC(C=1, kernel="rbf"),
    
]

for clf in classifiers:
    score = cross_validate(
        clf, 
        df[["x", "y","z"]],
        df["label"],
        scoring=["accuracy", "f1", "roc_auc"],
        cv=10
    )
    print("Mean Acc.:%2.4f +/- %1.2f F1-Score: %2.4f +/- %1.2f ROC-AUC: %2.4f" % (
        score["test_accuracy"].mean(), 
        score["test_accuracy"].std() * 2,
        score["test_f1"].mean(),
        score["test_f1"].std() * 2,
        score["test_roc_auc"].mean()
    ))
```

# Export as PMML

```python
from sklearn2pmml.pipeline import PMMLPipeline
from sklearn2pmml import sklearn2pmml
```

```python
pipeline = PMMLPipeline([
    RandomForestClassifier(n_estimators=5)
])
```

```python
pipeline.fit(
    df[["x", "y", "z"]],
    df["label"]
)
```

```python
sklearn2pmml(pipeline, "css18.pmml", with_repr = False)
```
